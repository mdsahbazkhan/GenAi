#### Using Pydantic for Validation

```python
from pydantic import BaseModel,Field
from typing import List,Annotated
```

```python
class User(BaseModel):
    name: str
    email: str
    age: int
```

```python
user=User(name="sahbaz", email="abc@gmail.com", age=30)
print(user)
```

#### Field Validation

```python
class Product(BaseModel):
    title: str =Field(max_length=15)
    description: str = Field(max_length=200)
    price: int = Field(gt=0)
```

```python
product1=Product(title="HP Laptops", description="A high-performance laptop for gaming and work.", price=1500)
product2=Product(title="Dell Laptop", description="A high-performance laptop for gaming and work.", price=2500)
# print(product)
```

```python
class Order(BaseModel):
    order_id: str =Field(min_length=5, max_length=10)
    product: Product
   
```

```python
order=Order(order_id="ORD123", product=product)
```

```python
order
```

```python
class Carts(BaseModel):
    id: str=Field(min_length=5, max_length=10)
    items: List[Product]
    comment:str = Field(default="No comment")
```

```python
cart = Carts(id="CART1123", items=[product1, product2], comment="This is my cart.")
```

```python
cart
```

```python
for item in cart.items:
    print(item.title)
```

```python
# Annotated - Custom validation
commonValidator = Annotated[List[Product],Field(default=[])]
```

```python
class Carts(BaseModel):
    id: str=Field(min_length=5, max_length=10)
    items: commonValidator
    comment:str = Field(default="No comment")
```

```python
class Order(BaseModel):
    product:commonValidator
```

```python
order=Order(product=[product1,product2])
```
