```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a:int,b:int):
    """
    It will Return the sum of two numbers.
    Args:
        a:First number
        b:Second number
    """
    return a+b

@tool
def multiply_numbers(a:int,b:int):
    """
    It will Return the product of two numbers.
    Args:
        a:First number
        b:Second number
    """
    return a*b
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite-preview",temperature=0.9)
```

```python
agent=create_agent(
    model=llm,
    tools=[add_numbers,multiply_numbers],
    system_prompt="You are a math teacher. Always use tools for any calculation."
)
```

```python
res=agent.invoke({"messages":[{"role":"user","content":"What is the 2+3*5 ?"}]})
print(res["messages"][-1].text)
```
