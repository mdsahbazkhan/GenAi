```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq
import os
llm=ChatGroq(
    model_name="llama-3.3-70b-versatile",
    temperature=0.7,
    groq_api_key=os.getenv("GROQ_API_KEY")
)
```

#### Discuss Problem First

```python
text="Hello, My name is Sahbaz Khan, " \
    "Email is amdsahbaz@gmail.com and My age is 30. " 
```

```python
res=llm.invoke(f"Please give me the name, email and age in json format from the following text: {text}")
print(type(res.content))
```

#### Solution: Structured Output

```python
from pydantic import BaseModel,Field
from typing import List
class ResponseStructured(BaseModel):
    name:str=Field(description="Name of the person")
    email:str=Field(description="Email of the person")
    age:int=Field(description="Age of the person")  

structured_llm=llm.with_structured_output(ResponseStructured)
```

```python
res=structured_llm.invoke(f"Please give me the name, email and age in json format from the following text: {text}")
res.model_dump()
```

```python
class Movie(BaseModel):
    title:str=Field(description="Title of the movie")
    director:str=Field(description="Director of the movie")
    year:int=Field(description="Release year of the movie")
    country:str=Field(description="Country of origin of the movie")
    
class AllMovies(BaseModel):
    movies:List[Movie]=Field(description="List of movies")

movie_llm=llm.with_structured_output(AllMovies)
```

```python
res=movie_llm.invoke("Give me five Drama movies of 2025")
res.model_dump()
```
