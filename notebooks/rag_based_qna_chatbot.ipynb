```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings,ChatGoogleGenerativeAI
from langchain_chroma import Chroma
from langchain_core.prompts import ChatPromptTemplate
```

```python
loader= PyPDFLoader("../data/GenAI_Course_Content.pdf")
docs= loader.load()
len(docs)
```

```python
splitter= RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
splitted_docs= splitter.split_documents(docs)
len(splitted_docs)
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-2")
```

```python
vector_store=Chroma.from_documents(
    documents=splitted_docs, 
    embedding=embeddings
    )
```

```python
llm= ChatGoogleGenerativeAI(model="gemini-3-flash-preview")
```

#### Chain - Context_generate | prompt | llm | strparser

```python
def get_context(query:str):
    data=vector_store.similarity_search(query=query)
    context=""
    for doc in data:
        context+=doc.page_content + "\n"
    return {
        "context": context,
        "question": query
    }
```

```python
prompt = ChatPromptTemplate.from_template("""
    You are helful assistent and provide answer based on the context for user question. And If you don't know the answer, 
    then you say "I don't know". 
    Context: {context}
    Question: {question}""")
```

```python
rag_chain= get_context | prompt | llm
```

```python
res=rag_chain.invoke("Can you me all module summary of GenAI course? And what will i learned and after this course,what skills i can add in my linkedin in short?")
```
