```python
from langchain_groq import ChatGroq
import os
llm = ChatGroq(
    model_name="llama-3.3-70b-versatile",
    temperature=0.7,
     api_key=os.getenv("GROQ_API_KEY")
)
```

```python
res=llm.invoke("Explain React in simple words")
print(res.content)
```
