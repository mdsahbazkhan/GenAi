```python
from langchain_groq import ChatGroq
import os
llm = ChatGroq(
    model_name="llama-3.3-70b-versatile",
    temperature=0.7,
     api_key=os.getenv("GROQ_API_KEY")
)
```

```python
prompts=[
    {"role": "user", "content": "Hello, My name is Sahbaz Khan"},
    {"role":"ai","content":"Hello Sahbaz Khan, it's nice to meet you. Is there something I can help you with or would you like to chat?"},
    {"role":"user","content":"Can you tell me my name?"},
]
res=llm.invoke(prompts)
print(res.content)
```

```python
history=[]

while True:
    query=input("You: ")
    if query.lower() in ["exit","quit","bye"]:
        print("Goodbye!")
        break
    history.append({"role":"user","content":query})
    print("You: ",query,"\n")
    
    res=llm.invoke(history)
    history.append({"role":"ai","content":res.content})
    print("AI: ",res.content,"\n")
```
