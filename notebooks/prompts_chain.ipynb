#### Prompt Templates - Dynamic Prompts

```python
from langchain_groq import ChatGroq
# from langchain_google_genai import ChatGoogleGenerativeAI
# llm=ChatGoogleGenerativeAI(model="gemini-3-flash-preview")
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers  import StrOutputParser
import os
llm = ChatGroq(
    model_name="llama-3.3-70b-versatile",
    temperature=0.7,
     api_key=os.getenv("GROQ_API_KEY")
) 
output_parser=StrOutputParser()  
```

```python
prompt= ChatPromptTemplate.from_messages([
    {"role":"system", "content":"You are a Translator and Tanslate input in {language}."},
    {"role":"user", "content":"{query}"}
])
```

#### Chains

```python
## My Self Chain Implementation

def transform_case(result:str):
    return result.upper()
```

```python
chains=prompt | llm | output_parser | transform_case
res=chains.invoke({
    "language": "English",
    "query": "میں لینگچین سے محبت کرتا ہوں۔"
})
print(res)
```
